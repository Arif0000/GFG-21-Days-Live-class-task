```python
# Import required libraries
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
# Step 1: Load GPT-2 Model and Tokenizer
model_name = "gpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Set model to evaluation mode
model.eval()


# Step 2: Function to check if the question is related to Python coding
def is_python_question(prompt):

    # List of keywords related to Python coding
    python_keywords = [
        "python", "code", "function", "loop", "list", "dictionary",
        "class", "object", "numpy", "pandas", "for loop", "while",
        "syntax", "variable", "print"
    ]

    prompt_lower = prompt.lower()

    for keyword in python_keywords:
        if keyword in prompt_lower:
            return True

    return False


# Step 3: Function to generate response using GPT-2
def generate_python_answer(prompt):

    inputs = tokenizer.encode(prompt, return_tensors="pt")

    outputs = model.generate(
        inputs,
        max_length=100,
        num_return_sequences=1,
        temperature=0.7,
        pad_token_id=tokenizer.eos_token_id
    )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return response


# Step 4: Main function that filters and generates response
def ask_model(prompt):

    if is_python_question(prompt):

        print("Python Coding Question Detected\n")

        response = generate_python_answer(prompt)

        return response

    else:
        return "Sorry, I can only answer Python coding questions."


# Step 5: Testing the system with different prompts

test_prompts = [
    "How do I write a for loop in Python?",
    "Explain Python list comprehension",
    "What is the capital of France?",
    "Write a Python function to add two numbers",
    "Who invented the telephone?"
]


for prompt in test_prompts:

    print("\nQuestion:", prompt)

    answer = ask_model(prompt)

    print("Answer:", answer)
```
